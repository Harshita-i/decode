```python
print('Outlier detection and removal using IQR - Method1')
```

```python
# calculate q1 (first quartile) and Q3(Third quartile).
q1=df.mpg.quantile(0.25)
q3=df.mpg.quantile(0.75)
```

```python
#Calculate IQR
IQR=q3-q1
IQR
```

```python
#calculate lowerlimit and upper limit
LL=q1-1.5*IQR
UL=q3+1.5*IQR
```

```python
UL
```

```python
LL
```

```python
#Upper bound 
upper = np.where(df['mpg'] >= UL)
upper
```

```python
#lower bound
lower = np.where(df['mpg']<=LL)
lower
```

```python
print('1. Dropping Outliers in upper and lower level of data')
```

```python
# delete the rows above upper and below lower values 
df.drop(upper[0], inplace=True)
df
```

```python
df.drop(lower[0], inplace=True)
df
```

```python
# indexing method 
# first convert the target values into an array 
arr= df2['acceleration'].values
arr
```

```python
# if arr values are in between ll and ul, represent that
# value as True else as False 
# False values are outliers we can find it at 5th position 
true_index = (arr>LL) & (arr < UL)
true_index
# Outliers are represented as False
```

```python
# let us substitute mean value in the place of outliers
# so ca1culate mean of all values other than the outliers.
mid1=np.mean(df2['acceleration'][true_index])
mid1
```

```python
#replace False index value by mean
false_index= ~true_index # ~ is bool operator for inverse
df2['acceleration'].values[false_index]=mid1
df2
```
